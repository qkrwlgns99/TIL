### 요소수 주유 유통은 031(경기도)에 밀집되어 있다

```python
# 주소를 활용한 분석
# split을 사용하여 첫 번째 값을 가져오는 extract함수 생성
def extract(string):
    return string.split()[0]
```

```python
addrs = df['addr'].apply(extract)
addrs
```

```python
addrs_cnt = addrs.value_counts().sort_values(ascending = False)
addrs_cnt
```

```python
addrs_cnt_df = pd.DataFrame(addrs_cnt).rename(columns = {'addr':'counts'})
addrs_cnt_df
```

```python
# 경기도와 경기, 인천광역시와 인천

def merge_rows(df, rowA, rowB):
    cnt = df.loc[rowB]
    df.loc[rowA] += cnt
    df = df.drop(rowB)
    return df
```

```python
addrs_cnt_df = merge_rows(addrs_cnt_df, '인천', '인천광역시')
addrs_cnt_df
```

```python
addrs_cnt_df = merge_rows(addrs_cnt_df, '경기', '경기도')
addrs_cnt_df
```

### 3-5 주유소 이름에서 패턴 찾아보기

```python
# 분석 키워드와 가격 차이를 입력값으로 받아, 전체 평균과 비교

def price_insight(keyword, diff):
    if diff > 0:
        print(f"{keyword}는 {diff}원 저렴")
    else:
        print(f"{keyword}는 {-diff}원 비쌈")
```

```python
# 분석 키워드, 재고량 차이를 입력 받아, 전체 평균과 비교

def inven_insight(keyword, diff):
    if diff > 0:
        print(f"{keyword}는 {diff} 적음")
    else:
        print(f"{keyword}는 {-diff} 많음")
```

```python
tdf = df[df.name.str.contains('휴게소')]
tdf
```

```python
price_diff = df.price.mean() - tdf.price.mean()
price_insight("휴게소", price_diff)
```

```python
inven_diff = df.inventory.mean() - tdf.inventory.mean()
inven_insight("휴게소", inven_diff)
```

```python
tdf = df[df.name.str.contains('셀프')]
print(tdf.shape[0])
tdf.head()
```

```python
price_diff = df.price.mean() - tdf.price.mean()
price_insight("셀프", price_diff)
```

```python
inven_diff = df.inventory.mean() - tdf.inventory.mean()
inven_insight("셀프", inven_diff)
```

```python
tdf = df[df.name.str.contains('알뜰')]
print(tdf.shape[0])
tdf.head()
```

```python
price_diff = df.price.mean() - tdf.price.mean()
price_insight("알뜰", price_diff)
```

### 4-1 지역별 요소수 재고 현황 지도에 나타내기
folium 라이브러리

```python
#folium 라이브러리 불러오기
!pip install folium
import folium
```

```python
# 중구를 가운데 좌표

map_osm = folium.Map(location = [37.557945, 126.99419], zoom_start = 10)
map_osm
```

```python
# 타겟 지역 설정
target_region = '경기'
```

```python
df_target_region = df[df['addr'].str.contains(target_region)]
df_target_region
```

```python
df_target_region.shape
```

1. 주유소 명칭, 주소, 재고량, 전화번호
2. lat, lng 정보 이용하여 지도 좌표를 가져온다
3. 재고에 따라 지도에 표출할 마커의 색상을 정한다
4. CircleMarker를 이용하여 마커를 생성

```python
map_osm = folium.Map(location=[37.557945, 126.99419], zoom_start = 9)

for i in df_target_region.values:
    iframe = "<b> 명칭 : </b>" + str(i[6]) + "<br><b>주소</b> : " + i[0] + "<br>" + "<b>재고량 : </b>" + str(i[3]) + "<br>" + "<b>전화번호 : </b>" + str(i[8])
    popup = folium.Popup(iframe, min_width=300, max_width=800)
    
    # 요소수 재고에 따라 나타낼 마커의 색상
    color = i[2].upper()
    if color == 'GREEN':
        color_code = '#9ACD32'
    elif color == 'YELLOW':
        color_code = '#FFFF00'
    elif color == 'RED':
        color_code = '#DC143C'
    elif color == 'GRAY':
        color_code = '#696969'
        
    # circleMarker
    marker = folium.CircleMarker([i[4], i[5]], 
                                 radius = 5,
                                 color = color_code,
                                 fill = True,
                                 fill_opacity = 1.0,
                                 popup = popup                      
                                )
    marker.add_to(map_osm)

map_osm
```

```python
# 가격에 대한 folium

map_osm = folium.Map(location=[37.557945, 126.99419], zoom_start = 9)

for i in df_target_region.values:
    iframe = "<b> 명칭 : </b>" + str(i[6]) + "<br><b>주소</b> : " + i[0] + "<br>" + "<b>재고량 : </b>" + str(i[3]) + "<br>" + "<b>전화번호 : </b>" + str(i[8])
    popup = folium.Popup(iframe, min_width=300, max_width=800)
    
    # 요소수 가격에 따라 나타낼 마커의 색상
    price = i[7]
    if price <= 1500:
        color_code = '#f9d71c'
    elif price <= 2000:
        color_code = '#ff6600'
    else:
        color_code = '#ff0000'
        
        
    # circleMarker
    marker = folium.CircleMarker([i[4], i[5]], 
                                 radius = 5,
                                 color = color_code,
                                 fill = True,
                                 fill_opacity = 1.0,
                                 popup = popup                      
                                )
    marker.add_to(map_osm)

map_osm
```
